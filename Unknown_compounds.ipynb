## Import Data and necessary packages for analyse

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.ticker as mticker
from matplotlib.colors import LogNorm


df1=pd.read_excel('/Users/nurali/Desktop/Filets unknowns compounds all.xlsx')
df2=pd.read_excel('/Users/nurali/Desktop/Filets compound classes All.xlsx')

data=df1.merge(df2, on='EPA Sample ID', how='left')

```

##  Showing data from 150 samples as below:

```python
data
```

## Pivot the data by different Lakes, filled the Nan value with : "0.001"

### Note: the value of compounds in each lake  was Mean concentration  not sum concentration


```python
lake_base=data.pivot_table("Concentration", "Lake", "Formula", aggfunc=np.mean)
lake_base=lake_base.fillna(0.001)
lake_base
```

# Heatmap plot the Unkowns data

## X axis is  compounds formula, Y axis is Lakes Name, Color changed by compounds concentration from dark to light (low to high), in Log scale

```python


sns.set(font_scale=1.5)


sns.clustermap(lake_base,metric="correlation",norm=LogNorm(lake_base.min().min(), lake_base.min().min()),cbar_kws={"shrink": .8,"ticks":[0,1,10,1e2,1e3]},vmin = 0.001, vmax=1000,figsize=(13, 13))




```

# Heatmap plot the Unkowns data's correlation

## X axis is  compounds formula, Y axis is compounds formula, Color changed by compounds correlation from dark to light (low to high)

## $Correlation$  $used$  $pearson$  $correlation $

```python
sns.set(font_scale=1.5)


sns.clustermap(lake_base.corr(),metric="correlation",figsize=(13, 13))

```

# Heatmap plot the Unkowns data's correlation

## X axis is  Lakes name, Y axis is also lakes name, Color changed by compounds correlation from dark to light (low to high)

## $Correlation$  $used$  $pearson$  $correlation $

```python


sns.set(font_scale=1.5)


lake_base=data.pivot_table("Concentration", "Formula","Lake", aggfunc=np.mean)
lake_base=lake_base.fillna(0.001)

g=sns.clustermap(lake_base.corr(),metric="correlation",figsize=(13, 13))

```
